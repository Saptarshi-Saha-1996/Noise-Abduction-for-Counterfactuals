```python
from cProfile import label
from turtle import position, title
from matplotlib import pyplot as plt
import seaborn as sns
plt.subplots(figsize=(10, 4))
sns.set_style('darkgrid')
#sns.set_style('whitegrid')
ax=sns.lineplot(data=time_train, x="sample_size", y="ratio",
     hue="batch_size", err_style='bars',palette=['red','black','green','maroon'], #"crest",#style="batch_size", #palette=['black','green'],
     legend="brief"
    )
#plt.tick_params(bottom=False,top=False, labelbottom=False)
ax.xaxis.tick_bottom()
plt.xticks([256,1024,4096,16384,32768],labels=['256','1024','4096','16384','32768'],rotation=45)
ax2=ax.secondary_xaxis('top')
ax2.set_xticks([512,8192],)#labels=['512'],rotation=45)
plt.ylabel("training time ratio",fontweight='bold')
plt.xlabel('sample size',fontweight='bold')
#ax.set_xticks_position('top')
#plt.xticklabels(['256','512','1024','4096','8192','16384','32768'],)
plt.legend(loc="best",ncol = 2,title='batch size',prop={'size': 9})
plt.savefig(experiment_name+'/training_time.pdf',format='pdf',pad_inches=0.1,bbox_inches='tight',dpi=1200)
```
